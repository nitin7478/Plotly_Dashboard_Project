```python
import pandas_datareader.data as web
import pandas as pd
import datetime as dt
import yfinance as yf
import datetime as dt
```

```python
Stocks = ['ITC.NS']
start = dt.datetime(1990,4,1)
end = dt.datetime(2023,2,27)
df = yf.download(Stocks, start=start, end=end)
object1 = yf.Ticker("ITC.NS")
object1.splits
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
scaler = MinMaxScaler(feature_range=(0,100))
x = df[['Close','Volume']].reset_index(drop =True)
x.columns
scaled_data = scaler.fit_transform(x)
scaled_data['Date'] = df.index
scaled_data.shape
```

```python
df1 = pd.DataFrame(scaled_data, columns=['Close','Volume'])
df1.index = df.index
```

```python
df1.head()
```

```python
import plotly.express as px
fig1 = px.line(x=df1.index.get_level_values('Date'), y=df1['Close'], title= 'Price-Volume Relation', labels= dict(x='Date', y='Close'))
fig1.add_bar(x=df1.index.get_level_values('Date'), y=df1['Volume'])
fig1.show()
```
